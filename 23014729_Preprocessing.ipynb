# Data Preprocessing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "loan_approval_dataset.csv"
df = pd.read_csv('/content/loan_approval_dataset.csv')
```

```python
# Trim whitespace from column names
df.columns = df.columns.str.strip()
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
```

```python
# Drop rows with missing values
df = df.dropna()
```

```python
# Display cleaned dataset
print(df.info())
print(df.head())
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Count plot of loan status
plt.figure(figsize=(6, 4))
sns.countplot(x=df['loan_status'], palette='viridis')
plt.title("Loan Status Distribution")
plt.xlabel("Loan Status")
plt.ylabel("Count")
plt.show()
```

```python
# Scatter plot of loan amount vs. CIBIL score
plt.figure(figsize=(8, 5))
sns.scatterplot(x='cibil_score', y='loan_amount', hue='loan_status', data=df, palette='coolwarm', alpha=0.7)
plt.title("Loan Amount vs. CIBIL Score by Loan Status")
plt.xlabel("CIBIL Score")
plt.ylabel("Loan Amount")
plt.show()
```
